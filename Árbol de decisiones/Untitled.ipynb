```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("./wine.data")
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
plt.hist(data.Class)
```

```python
predictors_vino  = ["Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
target_vino = ["Class"]
```

```python
predictors = data[predictors_vino]
target = data[target_vino]
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors,target, test_size=0.3, random_state=13)
```

```python
tree = DecisionTreeClassifier()

arbol = tree.fit(X_train, y_train)
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)

pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=["Actual"], colnames=["Predicciones"])
```

```python
acurracy = accuracy_score(y_test, predicciones)
acurracy
```
